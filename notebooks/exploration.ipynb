```python
import pandas as pd
import numpy as np

# G√©n√©rer 200 lignes
n = 10000

np.random.seed(42)  # pour que ce soit reproductible

data = {
    'id': np.arange(1, n+1),
    'age': np.random.randint(22, 61, size=n),
    'anciennete': np.random.randint(0, 34, size=n),
    'satisfaction': np.round(np.random.uniform(0.1, 1.0, size=n), 2),
    'salaire': np.random.randint(28000, 70000, size=n),
    'depart': np.random.randint(0, 2, size=n)  # 0 ou 1
}
data['promotions_recues'] = np.random.randint(0, 5, 10000)   # Entre 0 et 4 promotions
data['annees_depuis_derniere_promotion'] = np.random.randint(0, 10, 10000)
data['jours_absence'] = np.random.randint(0, 30, 10000)      # Nombre de jours d'absence
data['niveau_etude'] = np.random.choice(['Bac', 'Licence', 'Master', 'Doctorat'], 10000)
data['type_poste'] = np.random.choice(['Technicien', 'Manager', 'Directeur', 'Commercial'], 10000)
data['type_contrat'] = np.random.choice(['CDI', 'CDD'], 10000)
data['departement'] = np.random.choice(['RH', 'IT', 'Ventes', 'Finance', 'Production'], 10000)



df = pd.DataFrame(data)

# Afficher les 5 premi√®res lignes pour v√©rifier
print(df.head())

# Sauvegarder en CSV
df.to_csv('employes_rh.csv', index=False)
```
